```python
import pandas as pd 
train_df = pd.read_csv('/mnt/raid0/Kaggle/DS/new_byu-locating-bacterial-flagellar-motors-2025/train_labels.csv')
```

```python
df = train_df[train_df['Motor axis 0']!=-1]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame
df['ratio_0'] = df['Motor axis 0'] / df['Array shape (axis 0)']
df['ratio_1'] = df['Motor axis 1'] / df['Array shape (axis 1)']
df['ratio_2'] = df['Motor axis 2'] / df['Array shape (axis 2)']

# Set up the plot
plt.figure(figsize=(15, 5))

for i, ratio in enumerate(['ratio_0', 'ratio_1', 'ratio_2']):
    plt.subplot(1, 3, i + 1)
    sns.histplot(df[ratio], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {ratio}')
    plt.xlabel('Ratio')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
