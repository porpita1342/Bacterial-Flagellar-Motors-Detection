```python
pad_size = 1
input_size = 5
```

```python
exampler = torch.rand(11)
exampler
```

```python
p_exampler = F.pad(exampler, (1,1), "constant", 0)
p_exampler
```

```python
result = torch.tensor()
for x in range(0, p_exampler.size()[0], input_size - 2 * pad_size):
    print(x)
    if x + input_size > p_exampler.size()[0]:
        pad_amount = (x + input_size) - p_exampler.size()[0]
        print(F.pad(p_exampler[x:], (0, pad_amount), 'constant', 0))
        result = torch.cat(result, F.pad(p_exampler[x:], (0, pad_amount), 'constant', 0))
        break
    else:
        print(p_exampler[x:x + input_size])
    result = torch.cat(result,p_exampler[x:x + input_size])

result
```
